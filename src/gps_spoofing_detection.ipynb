```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
names=['0','1','2','lat','lon','alt','3','4','5','yaw','pitch','roll','6','x_mag','y_mag','z_mag','7','lin_acc_x','lin_acc_y','lin_acc_z']
```

```python
data = pd.read_csv('C:/Users/iampr/Downloads/major_project/dataset/TEST1.TXT', names=names)
```

```python
data.info()
```

```python
data.head(10)
```

```python
X = data.drop(labels = ['0','1','2','3','4','5','6','7'], axis = 1)
```

```python
X.head(10)
```

```python
r_num     = np.arange(len(X))
#gps
lat_data  = X['lat']
lon_data  = X['lon']
alt_data  = X['alt']
#3-axis gyroscope
y_data    = X['yaw']
p_data    = X['pitch']
r_data    = X['roll']
#3-axis magnetometer
x_mag     = X['x_mag']
y_mag     = X['y_mag']
z_mag     = X['z_mag']
#3-axis accelerometer
lin_acc_x = X['lin_acc_x']
lin_acc_y = X['lin_acc_y']
lin_acc_z = X['lin_acc_z']
```
